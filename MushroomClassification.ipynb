First Analysis Of Dataset and Looking its Tables

Dataset can be found at kaggle.com

```python
import pandas as pd
data = pd.read_csv('mushrooms.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

According to Dataset Analysis Delete Some Unneccessary Tables

```python
data.drop('veil-type', axis=1,inplace=True)
data.drop('stalk-root', axis=1,inplace=True)
data.info()
```

```python
# target columns possible values
data['class'].unique()
```

Now convert string columns to numbers to analyze better

```python
from sklearn.preprocessing import LabelEncoder
# this encoder is used to convert our dataframe values to the numeric equals
# fit() and transform() are same with fit_transform()
le = LabelEncoder()
for i in data.columns:
    data[i] = le.fit_transform(data[i])
data.head()
```

```python
data.describe()
```

Labels and Features to Estimate

```python
# Y is our target columns to predict
Y = data['class']
data1 = data
data1.drop('class', axis=1, inplace=True)
data1.head()
```

```python
# now data1 is my training and test set
X = data1
# now divide X into training and test set by train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, random_state=42, test_size=0.33)
```

Model Selection for Our Dataset

```python
# tried logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
model.score(X_test, Y_test)
```

```python
# try another model Support Vector Macihens
from sklearn import svm
model2 = svm.SVC(kernel='rbf', gamma=0.3)
model2.fit(X_train,Y_train)
model2.score(X_test,Y_test)
```

```python
# With LinearRegression
from sklearn.linear_model import LinearRegression
model3 = LinearRegression()
model3.fit(X_train,Y_train)
model3.score(X_test,Y_test)
```
